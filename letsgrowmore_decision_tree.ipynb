#### Importing Labraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
#%matplotlib imline
```

#### Loading Dataset

```python
import sklearn.datasets as datasets
iris_set = datasets.load_iris()
iris_data= pd.DataFrame(iris_set.data, columns=iris_set.feature_names)
iris_data["Species"] = iris_set.target
iris_data.head()
```

#### Visualization of Dataset 

```python
sns.set_style("darkgrid")
plt.figure(figsize=(12, 6))
plt.title('Sepal Dimensions')
sns.scatterplot(x=iris_data["sepal length (cm)"], 
                y=iris_data["sepal width (cm)"], 
                hue=iris_data["Species"],palette = ["green","orange","blue"],
                s=100);
```

```python
plt.figure(figsize=(12, 6))
plt.title('Petal Dimensions')

sns.scatterplot(x=iris_data["petal length (cm)"], 
                y=iris_data["petal width (cm)"], 
                hue=iris_data["Species"],palette = ["green","orange","blue"],
                s=100);
```

#### Spliting Dataset 

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(iris_data, test_size = 0.25)
```

```python
train.shape, test.shape
```

```python
train_x = train[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)','petal width (cm)']]
train_y = train.Species
test_x = test[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)','petal width (cm)']]
test_y = test.Species
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
dtree = DecisionTreeClassifier()
dtree.fit(train_x,train_y)
predictions = dtree.predict(test_x)
print("The accuracy of Decision Tree is:", accuracy_score(predictions, test_y))
```

```python
X = iris_data[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)','petal length (cm)']]
Y = iris_data.Species
```

```python
tree1 = DecisionTreeClassifier()
tree2=tree1.fit(X,Y)
y_pred=tree1.predict(test_x)
```

```python
print("Trainig Score: ",tree1.score(train_x,train_y))
```

```python
print(classification_report(test_y,y_pred))
```

```python
print(confusion_matrix(test_y,y_pred))
```

```python
print(accuracy_score(test_y,y_pred))
```
